```python
# notebook setup and import statements

%pylab inline
from JSAnimation.IPython_display import display_animation
import math
from matplotlib import animation, cm
import random
```

```python
# data for basic plotting

xInitial = np.random.randn(200)
yInitial = np.random.randn(200)
```

```python
# basic plotting

fig = plt.figure(0)
ax = fig.gca()

ax.scatter(xInitial, yInitial)
plt.show()
```

```python
# data for explosion animation

# some constants for the "explosion"
speed = .1
noise_magnitude = speed/2
timeSteps = 90
displacementMagnitude = np.sqrt(xInitial**2 + yInitial**2)
xFlow = xInitial/displacementMagnitude
yFlow = yInitial/displacementMagnitude

# simulation of the particle trajectories during the "explosion"
xTrajectory = np.zeros((200, timeSteps))
yTrajectory = np.zeros((200, timeSteps))
xTrajectory[:,0] = xInitial
yTrajectory[:,0] = yInitial
for i in range(1,timeSteps):
    xTrajectory[:,i] = xTrajectory[:,i-1] + \
                       xFlow*speed + \
                       np.random.uniform(-noise_magnitude, noise_magnitude, 200)
    yTrajectory[:,i] = yTrajectory[:,i-1] + \
                       yFlow*speed + \
                       np.random.uniform(-noise_magnitude, noise_magnitude, 200)
```

```python
# animating the plot

# declare the matplotlib figure object
fig = plt.figure(0)

# some constants for the animation
frames = 90  # same as timeSteps
fps = 60

# the animate function, which gets called once per frame of your animation
def animate(frame):
    fig.clf()
    ax = fig.gca()
    ax.scatter(xTrajectory[:,frame], yTrajectory[:,frame])

anim = animation.FuncAnimation(fig, animate, frames=frames)
display_animation(anim, default_mode='loop', fps=fps)
```

```python
# labeling and prettying up the animation

# declare the matplotlib figure object
fig = plt.figure(0)

# some constants for the animation
frames = 90  # same as timeSteps
fps = 60

# the animate function, which gets called once per frame of your animation
def animate(frame):
    fig.clf()
    fig.set_size_inches(12,8)
    ax = fig.gca()
    ax.scatter(xTrajectory[:,frame], yTrajectory[:,frame])
    
    # fix up and label the X axis
    ax.set_xlim(-10,10)
    ax.set_xlabel('the X axis', fontsize=30)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(30)
    
    # fix up and label the Y axis
    ax.set_ylim(-10,10)
    ax.set_ylabel('the Y axis', fontsize=30)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(30)

anim = animation.FuncAnimation(fig, animate, frames=frames)
display_animation(anim, default_mode='loop', fps=fps)

# if you want to save the animation to file rather than display it, uncomment one of the following lines
#anim.save('random_scatter_animation.mp4',writer='ffmpeg', fps=30)
#anim.save('random_scatter_animation.gif',writer='imagemagick', fps=30)
```

```python
# adding an animated label to the labeled animation

# declare the matplotlib figure object
fig = plt.figure(0)

# some constants for the animation
frames = 90  # same as timeSteps
fps = 60

# the animate function, which gets called once per frame of your animation
def animate(frame):
    fig.clf()
    fig.set_size_inches(12,8)
    ax = fig.gca()
    ax.scatter(xTrajectory[:,frame], yTrajectory[:,frame])
    
    # fix up and label the X axis
    ax.set_xlim(-10,10)
    ax.set_xlabel('the X axis', fontsize=30)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(30)
    
    # fix up and label the Y axis
    ax.set_ylim(-10,10)
    ax.set_ylabel('the Y axis', fontsize=30)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(30)
    
    # a text object that gets updated once per frame, and so appears animated
    ax.text(0.7, 0.9, 'frame = %d' % frame, 
            horizontalalignment='left', verticalalignment='bottom', 
            fontsize=30, transform = ax.transAxes)

anim = animation.FuncAnimation(fig, animate, frames=frames)
display_animation(anim, default_mode='loop', fps=fps)
```

```python
# giving the particles random shapes and rainbow colors (ie. glitterBomb)

# declare the matplotlib figure object
fig = plt.figure(0)

# some constants for the animation
frames = 90  # same as timeSteps
fps = 60

# define a list of matplotlib scatter markers to choose from
markers = ['.', ',', 'o', 'v', '^', '>', '<', '1', '2', '3', '4', '8', 's', 'p', '*'] 
# chose a marker and a color for each particle in the explosion
colorMarkers = []
for i in range(200):
    colorMarkers.append((cm.jet(np.random.rand()), random.choice(markers)))

# the animate function, which gets called once per frame of your animation
def animate(frame):
    fig.clf()
    fig.set_size_inches(12,8)
    ax = fig.gca()
    # instead of a normal scatterplot, we make a new plot (using a different color and marker) for each data point
    for i, (color, marker) in enumerate(colorMarkers):
        ax.plot(xTrajectory[i,frame], yTrajectory[i,frame], c=color, marker=marker, markersize=10)
    
    # fix up and label the X axis
    ax.set_xlim(-10,10)
    ax.set_xlabel('the X axis', fontsize=30)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(30)
    
    # fix up and label the Y axis
    ax.set_ylim(-10,10)
    ax.set_ylabel('the Y axis', fontsize=30)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(30)

        
anim = animation.FuncAnimation(fig, animate, frames=frames)
display_animation(anim, default_mode='loop', fps=fps)
```
