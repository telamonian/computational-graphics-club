```python
# animating the plot
from matplotlib import animation
 
fig = plt.figure(0)
frames = 5
xInitial = np.random.randn(200)
yInitial = np.random.randn(200)

def animate(frame):
    fig.clf()
    ax = fig.gca()
    xCurrent = xInitial + np.cos(frame/float(frames)*math.pi*2)
    yCurrent = yInitial + np.sin(frame/float(frames)*math.pi*2)
    ax.scatter(xCurrent, yCurrent)

anim = animation.FuncAnimation(fig, animate, frames=frames)
anim.save('random_scatter_animation.mp4',writer='ffmpeg', fps=30);
```

```python
# labeling and prettying up the animation
from matplotlib import animation
 
fig = plt.figure(0)
frames = 65
xInitial = np.random.randn(200)
yInitial = np.random.randn(200)

def animate(frame):
    fig.clf()
    fig.set_size_inches(12,8)
    ax = fig.gca()
    xCurrent = xInitial + np.cos(frame/float(frames)*math.pi*2)
    yCurrent = yInitial + np.sin(frame/float(frames)*math.pi*2)
    ax.scatter(xCurrent, yCurrent)
    
    ax.set_xlim(-5,5)
    ax.set_xlabel('the X axis', fontsize=30)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(30)
    
    
    ax.set_ylim(-5,5)
    ax.set_ylabel('the Y axis', fontsize=30)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(30)
    
anim = animation.FuncAnimation(fig, animate, frames=frames)
anim.save('random_scatter_animation.gif',writer='imagemagick', fps=30);
```

```python
# adding an animated label to the labeled animation
from matplotlib import animation
 
fig = plt.figure(0)
frames = 65
xInitial = np.random.randn(200)
yInitial = np.random.randn(200)

def animate(frame):
    fig.clf()
    fig.set_size_inches(12,8)
    ax = fig.gca()
    xCurrent = xInitial + np.cos(frame/float(frames)*math.pi*2)
    yCurrent = yInitial + np.sin(frame/float(frames)*math.pi*2)
    ax.scatter(xCurrent, yCurrent)
    
    ax.set_xlim(-5,5)
    ax.set_xlabel('the X axis', fontsize=30)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(30)
    
    ax.set_ylim(-5,5)
    ax.set_ylabel('the Y axis', fontsize=30)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(30)
    
    ax.text(2.2, 4, 'frame = %d' % frame, fontsize=30)
    
anim = animation.FuncAnimation(fig, animate, frames=frames)
anim.save('random_scatter_animation.gif',writer='imagemagick', fps=30);
```
